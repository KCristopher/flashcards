```python
df = pd.read_csv('https://raw.githubusercontent.com/KCristopher/flashcards/main/data/french_words.csv')
df.head()
```

```python
words_translations = [*zip(df['French'], df['in English'])]
words_translations
```

```python
unique_words = {f for f in df.French}
translations_by_words = {}

for u in unique_words :
    translations_by_words[u] = df[df.French == u]['in English'].values.tolist()
```

```python
{k : v for k, v in translations_by_words.items() if len(v) > 2}
```

```python
r.sample(words_translations, 10)
```

```python
answers_by_word = {}

for word, translation in words_translations[ : 5] :
    print(word)
    t.sleep(1.5)
    reveal = input('Reveal the translation? (y/n)').strip().lower()
    if reveal == 'y' :
        print(translation)
        t.sleep(1)
        did_you_know = input('Did you know the translation? (y/n/mispronounce)').strip().lower()
        answers_by_word[word] = did_you_know
```

```python
incorrect_answers = [w for w, a in answers_by_word.items() if a == 'n']
incorrect_answers
```
